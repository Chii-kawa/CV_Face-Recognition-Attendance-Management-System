```python
import os

root = 'data/k-face/new_path/rename_train_align_112'

for path, subdirs, files in os.walk(root):
    print(f"Current directory: {path}")
    for i, name in enumerate(files):
        print(f"Original filename: {name}")
        
        extension = name.split(".")[-1].lower()
        print(f"File extension: {extension}")
        
        if extension != "jpg":
            print(f"Skipping file: {name} (Not a .jpg file)")
            continue
        
        # Construct new filename with zero padding to 4 digits
        new_name = f'{os.path.basename(path)}_{str(i + 1).zfill(4)}.{extension}'
        print(f"New filename: {new_name}")
        
        # Rename the file
        try:
            old_file = os.path.join(path, name)
            new_file = os.path.join(path, new_name)
            print(f"Renaming {old_file} to {new_file}")
            os.rename(old_file, new_file)
        except Exception as e:
            print(f"Error renaming file {name} to {new_name}: {e}")
        
    print(f"Processed directory: {os.path.basename(path)}")
```
